```python
import pandas as pd
import os

INTER_DIR = "../data/02_intermediate/"
DOCS_DIR = "../docs/"
os.makedirs(DOCS_DIR, exist_ok=True)

print("Vorbereitung Identity Resolution...")

# Gemappte Zwischendaten laden
df_pks = pd.read_csv(os.path.join(INTER_DIR, 'pks_mapped.csv'), sep=';')
df_jus = pd.read_csv(os.path.join(INTER_DIR, 'justiz_mapped.csv'), sep=';')

# Eindeutige Straftat-Bezeichnungen beider Quellen extrahieren
pks_straftaten = df_pks['straftat_bezeichnung'].dropna().unique()
jus_straftaten = df_jus['straftat_bezeichnung'].dropna().unique()

print(f"PKS-Delikte:       {len(pks_straftaten)}")
print(f"Justiz-Kategorien: {len(jus_straftaten)}")

# Justiz-Kategorien als Referenz-CSV exportieren
df_jus_ref = pd.DataFrame({'justiz_kategorie': sorted(jus_straftaten)})
df_jus_ref.to_csv(os.path.join(DOCS_DIR, 'justiz_kategorien_referenz.csv'), index=False, sep=';')

# Leere Mapping-Tabelle erstellen (PKS-Delikt -> Justiz-Kategorie)
# Die Zuordnung erfolgt in Notebook 05 automatisiert + manuelle Nacharbeit
df_mapping = pd.DataFrame({
    'pks_straftat_bezeichnung': sorted(pks_straftaten),
    'justiz_straftat_bezeichnung_mapped': ''
})

mapping_path = os.path.join(DOCS_DIR, 'mapping_table.csv')
df_mapping.to_csv(mapping_path, index=False, sep=';')

print(f"Mapping-Tabelle erstellt: {mapping_path}")
```
